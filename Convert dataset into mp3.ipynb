```python
import pandas
```

```python
main_file_path = 'dataset/main_song_labels.csv'
song_dir = 'dataset/song_preview'
song_df = pandas.read_csv(main_file_path, sep='\t')
song_df.head()
```

```python
import subprocess
import os
from tqdm import tqdm

OUTPUT_DIR = 'dataset/song_mp3'
song_df['mp3_file']=''

for idx,row in tqdm(song_df.iterrows(), 'converting to m4a dataset into mp3:'):
    filename = row['preview_file']
    output = os.path.join(OUTPUT_DIR, '%s.mp3' % filename[:-4])
    subprocess.call([
        "ffmpeg", "-i",
        os.path.join(song_dir, filename),
        "-acodec", "libmp3lame", "-ab", "256k",
        output
        ])
    song_df.mp3_file.loc[idx] = output
song_df.head(10)
```
